```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r'D:\CodSoft\DS\Iris-Flower-Classification\IRIS.csv')
df.head()
```

```python
df['species'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['species'] = le.fit_transform(df['species'])
```

```python
df.info()
```

```python
X = df.drop('species', axis = 1)
y = df['species']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 10)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.linear_model import RidgeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.linear_model import Perceptron
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.ensemble import StackingClassifier
from sklearn.gaussian_process import GaussianProcessClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.model_selection import cross_val_score

models = [
    LogisticRegression(),
    SVC(),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    KNeighborsClassifier(),
    GaussianNB(),
    QuadraticDiscriminantAnalysis(),
    RidgeClassifier(),
    GradientBoostingClassifier(),
    AdaBoostClassifier(),
    BaggingClassifier(),
    Perceptron(),
    HistGradientBoostingClassifier(),
    GaussianProcessClassifier(),
    XGBClassifier(),
    LGBMClassifier(),
    CatBoostClassifier(silent=True)
]

for model in models:
    model_name = model.__class__.__name__
    
    # Fit the model on the training data
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    # Calculate accuracy score on the test set
    accuracy = accuracy_score(y_test, y_pred)
    
    # Perform cross-validation and calculate the mean score
    cross_val = cross_val_score(model, X_train, y_train, cv=5, scoring='accuracy').mean()
    
    print(f"{model_name} Accuracy: {accuracy:.4f} | Cross-Val Score: {cross_val:.4f}")
```

```python
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings("ignore")  # Suppress warnings for cleaner output

# Define models and corresponding parameter grids
model_param_grid = {
    'LogisticRegression': {
        'model': LogisticRegression(),
        'params': {'C': [0.01, 0.1, 1, 10], 'solver': ['liblinear', 'lbfgs']}
    },
    'SVC': {
        'model': SVC(),
        'params': {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf']}
    },
    'DecisionTreeClassifier': {
        'model': DecisionTreeClassifier(),
        'params': {'criterion': ['gini', 'entropy'], 'max_depth': [None, 10, 20, 30]}
    },
    'RandomForestClassifier': {
        'model': RandomForestClassifier(),
        'params': {'n_estimators': [50, 100, 200], 'max_depth': [None, 10, 20]}
    },
    'KNeighborsClassifier': {
        'model': KNeighborsClassifier(),
        'params': {'n_neighbors': [3, 5, 7], 'weights': ['uniform', 'distance']}
    },
    'GaussianNB': {
        'model': GaussianNB(),
        'params': {}  # No hyperparameters for GaussianNB
    },
    'QuadraticDiscriminantAnalysis': {
        'model': QuadraticDiscriminantAnalysis(),
        'params': {}  # No hyperparameters for QDA
    },
    'RidgeClassifier': {
        'model': RidgeClassifier(),
        'params': {'alpha': [0.01, 0.1, 1, 10]}
    },
    'GradientBoostingClassifier': {
        'model': GradientBoostingClassifier(),
        'params': {'n_estimators': [50, 100, 200], 'learning_rate': [0.01, 0.1, 0.2]}
    },
    'AdaBoostClassifier': {
        'model': AdaBoostClassifier(),
        'params': {'n_estimators': [50, 100, 200], 'learning_rate': [0.01, 0.1, 1]}
    },
    'BaggingClassifier': {
        'model': BaggingClassifier(),
        'params': {'n_estimators': [10, 50, 100]}
    },
    'Perceptron': {
        'model': Perceptron(),
        'params': {'penalty': [None, 'l2', 'l1'], 'alpha': [0.0001, 0.001, 0.01]}
    },
    'HistGradientBoostingClassifier': {
        'model': HistGradientBoostingClassifier(),
        'params': {'learning_rate': [0.01, 0.1, 0.2], 'max_iter': [100, 200]}
    },
    'GaussianProcessClassifier': {
        'model': GaussianProcessClassifier(),
        'params': {}  # Limited parameter grid for simplicity
    },
    'XGBClassifier': {
        'model': XGBClassifier(),
        'params': {'learning_rate': [0.01, 0.1, 0.2], 'n_estimators': [50, 100, 200]}
    },
    'LGBMClassifier': {
        'model': LGBMClassifier(),
        'params': {'learning_rate': [0.01, 0.1, 0.2], 'n_estimators': [50, 100, 200]}
    },
    'CatBoostClassifier': {
        'model': CatBoostClassifier(silent=True),
        'params': {'learning_rate': [0.01, 0.1, 0.2], 'depth': [4, 6, 8]}
    }
}

# Perform hyperparameter tuning
best_models = {}
for name, mp in model_param_grid.items():
    print(f"Tuning hyperparameters for {name}...")
    grid = GridSearchCV(mp['model'], mp['params'], cv=5, scoring='accuracy', n_jobs=-1)
    grid.fit(X_train, y_train)
    best_models[name] = grid.best_estimator_
    print(f"Best Params for {name}: {grid.best_params_}")
    print(f"Best CV Score for {name}: {grid.best_score_:.4f}\n")
```

```python
from sklearn.model_selection import cross_val_score

# Initialize models with their best parameters
models = [
    LogisticRegression(C=10, solver='lbfgs'),
    SVC(C=1, kernel='linear'),
    DecisionTreeClassifier(criterion='gini', max_depth=None),
    RandomForestClassifier(n_estimators=50, max_depth=None),
    KNeighborsClassifier(n_neighbors=3, weights='uniform'),
    GaussianNB(),
    QuadraticDiscriminantAnalysis(),
    RidgeClassifier(alpha=0.01),
    GradientBoostingClassifier(learning_rate=0.1, n_estimators=100),
    AdaBoostClassifier(learning_rate=0.01, n_estimators=50),
    BaggingClassifier(n_estimators=10),
    Perceptron(alpha=0.0001, penalty='l2'),
    HistGradientBoostingClassifier(learning_rate=0.1, max_iter=100),
    GaussianProcessClassifier(),
    XGBClassifier(learning_rate=0.01, n_estimators=50),
    LGBMClassifier(learning_rate=0.1, n_estimators=100),
    CatBoostClassifier(learning_rate=0.01, depth=4, silent=True)
]

# Evaluate each model
for model in models:
    model_name = model.__class__.__name__
    
    # Fit the model on the training data
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    # Calculate accuracy score on the test set
    accuracy = accuracy_score(y_test, y_pred)
    
    # Perform cross-validation and calculate the mean score
    cross_val = cross_val_score(model, X_train, y_train, cv=5, scoring='accuracy').mean()
    
    print(f"{model_name} Accuracy: {accuracy:.4f} | Cross-Val Score: {cross_val:.4f}")
```

```python
base_models = [
    ('lr', LogisticRegression(C=10, solver='lbfgs')),
    ('svc', SVC(C=1, kernel='linear', probability=True)),
    ('knn', KNeighborsClassifier(n_neighbors=3, weights='uniform')),
    ('lgbm', LGBMClassifier(learning_rate=0.02, n_estimators=100))
]

meta_model = LogisticRegression(C=10, solver='liblinear')

stacking_model = StackingClassifier(estimators=base_models, final_estimator=meta_model, cv=3)

stacking_model.fit(X_train, y_train)

y_pred = stacking_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

cross_val = cross_val_score(stacking_model, X_train, y_train, cv=3, scoring='accuracy').mean()

print(f"Stacking Model Accuracy: {accuracy:.4f}")
print(f"Stacking Model Cross-Val Score: {cross_val:.4f}")
```

```python
param_grid = {
    'final_estimator__C': [0.01, 0.1, 1, 10, 100],
    'final_estimator__solver': ['lbfgs', 'liblinear'],
    'cv': [3, 5, 7]
}

grid_search = GridSearchCV(estimator=stacking_model, param_grid=param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train, y_train)

best_stacking_model = grid_search.best_estimator_

y_pred = best_stacking_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
cross_val = cross_val_score(best_stacking_model, X_train, y_train, cv=5, scoring='accuracy').mean()

print(f"Best Params for Stacking Model: {grid_search.best_params_}")
print(f"Stacking Model Accuracy: {accuracy:.4f}")
print(f"Stacking Model Cross-Val Score: {cross_val:.4f}")
```
